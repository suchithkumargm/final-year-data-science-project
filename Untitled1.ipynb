```python
import csv
import pandas as pd
from selenium import webdriver

def fetch_image_url(website_url, image_xpath):
    try:
        # Create a new instance of the Chrome driver
        driver = webdriver.Chrome()
        
        # Open the webpage
        driver.get(website_url)
        
        # Find the image element using XPath
        image_element = driver.find_element("xpath", image_xpath)
        
        # Get the URL of the image
        image_url = image_element.get_attribute("src")
        
        return image_url
    
    except Exception as e:
        # Handle any errors
        print(f"Error fetching image URL for {website_url}: {str(e)}")
        return None
    
    finally:
        # Close the browser window
        driver.quit()

# Read the existing CSV file
csv_file = "Images_division/Output121.csv"

# Create an empty DataFrame to store the results
output_data = []

# Process URLs and store the results in the DataFrame
with open(csv_file, mode='r', newline='') as infile:
    reader = csv.reader(infile)
    
    try:
        # Read the header row
        header = next(reader)
    except StopIteration:
        print("The input CSV file is empty.")
        exit()

    # Process URLs in batches
    batch_size = 100  # Adjust batch size as needed
    for row in reader:
        website_url = row[1]  # Assuming website URL is in the second column
        image_xpath = '//*[@id="root"]/div/main/div/section[2]/div[1]/div/div/img'  # Replace with the actual XPath of your image
        image_url = fetch_image_url(website_url, image_xpath)
        
        if image_url is not None:
            output_data.append([website_url, image_url])

            if len(output_data) >= batch_size:
                # Create a DataFrame from the output data
                output_df = pd.DataFrame(output_data, columns=['Website URL', 'Image URL'])

                # Write the DataFrame to a new CSV file
                output_csv_file = "output_website_url_121.csv"
                output_df.to_csv(output_csv_file, index=False)

                print(f"Batch of image URLs have been written to {output_csv_file}")

                output_data = []  # Clear the output data for the next batch

# Write any remaining rows with image URLs to the same CSV file
if output_data:
    output_df = pd.DataFrame(output_data, columns=['Website URL', 'Image URL'])
    output_df.to_csv(output_csv_file, mode='a', index=False, header=False)

    print(f"Remaining image URLs have been written to {output_csv_file}")
```
